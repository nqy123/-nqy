打印图片:np.random.randint(low, high=None, size=None, dtype=int)
low：该参数是必需的，它指定了随机整数范围的下限（包含该下限值）。
high：这是一个可选参数，若指定了该参数，它表示随机整数范围的上限（不包含该上限值）。若不指定，默认从 0 到 low 这个区间生成随机整数，此时 low 代表上限（不包含该上限值）。
size：同样是可选参数，用于指定生成随机整数数组的形状。如果不指定，函数将返回一个单个的随机整数；若传入整数，则返回指定长度的一维数组；若传入元组，则返回对应形状的多维数组。

np.random.choice(a, size=None, replace=True, p=None)a：这是必需参数，表示用于抽样的一维数组或者整数 n。若为整数 n，则函数会将其视为 np.arange(n) 进行抽样。size：可选参数，用于指定输出的样本数组的形状。若不指定，默认返回单个元素；若传入整数，则返回对应长度的一维数组；若传入元组，则返回对应形状的多维数组。
images[10*r+c]这里假设子图布局是 10 行 10 列的网格结构，r 表示当前子图所在的行索引（从 0 到 9），c 表示当前子图所在的列索引（从 0 到 9）。10*r+c 这个计算方式是将二维的行和列索引转换为一维的索引，从而从 images 数组中选取对应的图像数据
前 r 行一共有 10 * r 个图像。
再加上当前第 r 行的第 c 列的偏移量 c，就得到了该子图对应的图像在 images 数组中的一维索引 10*r+c。

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# 加载数据
data = sio.loadmat(r"C:\Users\HP\Desktop\高速下载-17457248894892024吴恩达资料\2014版 吴恩达机器学习全套资料\吴恩达机器学习课后作业源码及笔记\ex3-neural network\ex3data1.mat")
x = data['X']  # 特征数据 (5000, 400)
y = data['y'].ravel()  # 转换为一维数组
y[y == 10] = 0  # 调整标签，将10改为0

# 随机展示单个图像
def dytp(x):
    pickone = np.random.randint(5000)
    image = x[pickone, :]
    fig, ax = plt.subplots(figsize=(1, 1))
    ax.imshow(image.reshape(20, 20).T, cmap='gray_r')
    plt.axis('off')
    return pickone

# 随机展示100个图像
def dytp100(x):
    pick100 = np.random.choice(5000, 100)
    images = x[pick100, :]
    fig, ax = plt.subplots(10, 10, figsize=(6, 6), sharex=True, sharey=True)
    for r in range(10):
        for c in range(10):
            ax[r, c].imshow(images[10*r + c].reshape(20, 20).T, cmap='gray_r')
            ax[r, c].axis('off')
    plt.show()
    return pick100

# 特征标准化
def pf(x):
    return (x-x.mean())/x.std()
x=pf(x)

# 划分训练集和测试集
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y
)

# 创建并训练逻辑回归模型
l1 = LogisticRegression(
    multi_class='ovr',
    solver='liblinear',
    C=1.0,
    max_iter=1000,
    random_state=42
)
l1.fit(x_train, y_train)

# 模型评估
y_trp = l1.predict(x_train)
y_tep= l1.predict(x_test)
print('训练集准确率', np.mean(y_trp==y_train))
print('测试集准确率' ,  np.mean(y_tep==y_test))

# 单个图像预测
print("\n单个图像预测:")
idx = dytp(x)
impf = pf(x[idx:idx+1, :])
prediction = l1.predict(impf)
print("预测结果:",prediction[0], "实际标签:",y[idx])

# 100个图像预测
print("\n100个图像预测:")
indices = dytp100(x)
impf = pf(x[indices, :])
predictions = l1.predict(impf)
acc = y[indices]
print("100个样本的预测准确率:", np.mean(acc==predictions))
```
